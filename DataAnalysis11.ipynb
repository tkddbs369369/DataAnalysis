```python
import csv

f = open('subwaytime.csv')
data = csv.reader(f)
next(data)
next(data)

mx = [0] * 24
mx_sta = [''] * 24

for row in data:
    row[4:] = map(int, row[4:])
    for i in range(24):
        a = row[i*2+4]
        if a > mx[i]:
            mx[i] = a
            mx_sta[i] = row[3]

print(mx_sta,mx)
```

```python
import matplotlib.pyplot as plt
plt.rc('font', family='malgun Gothic')
plt.bar(range(24),mx)

plt.xticks(range(24), mx_sta,rotation=90)
```

```python
import csv
import matplotlib.pyplot as plt
f = open('subwaytime.csv')
data = csv.reader(f)

next(data)
next(data)

s_in = [0] *24
s_out = [0] *24

for row in data:
    row[4:] = map(int, row[4:])
    for j in range(24):
        s_in[j] += row[4 + j*2]
        s_out[j] += row[5 + j*2]

plt.rc('font', family = 'Malgun Gothic')
plt.title('지하철 시간대별 승하차 인원 추이')
plt.plot(s_in, label = '승차')
plt.plot(s_out, label = '하차')
plt.legend()
plt.xticks(range(24), range(4,28))
plt.show()
```
