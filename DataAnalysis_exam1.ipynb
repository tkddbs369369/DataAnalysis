#### 5. age.csv 파일에서 지역명을 입력받아 인구 구조를 꺾은선 그래프로 출력하세요. (30점)

인구 구조가 알고 싶은 지역의 이름(읍면동 단위)을 입력해주세요 : 장위제1동


````python
import csv
import matplotlib.pyplot as plt
row = csv.reader(open('age.csv'))
next(row)
ad = input('원하는 동내')
re = []

for i in row:

    if ad in i[0]:
        for a in i[3:]:
            re.append(float(a))

plt.style.use('ggplot')
plt.rc('font', family='Malgun Gothic')
plt.title(ad+'의 지역구조')
plt.plot(re,'r')
````
