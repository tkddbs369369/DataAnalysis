```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

high = [] #     high 
low = [] #     low 

for row in data:
    if row[-1] != '' and row[-2] != "":
        date = row[0].split('-')
        if 1983 <= int(data[0]):
            if date[1] == '02' and date[2] == "15":
                high.append(float(row[-1]))
                low.append(float(row[-2]))


plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('    ')

plt.plot(high, 'hotpink', label = '')
plt.plot(low, 'skyblue', label = '')

plt.legend()
```
