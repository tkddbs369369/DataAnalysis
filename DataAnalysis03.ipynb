## 2월의 최고 기온 데이터를 추출하여 그래프로 그리기

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

result =list()

for i in data:
    if i[-1] !="":
        if i[0].split('-')[1] == '02':
            result.append(float(i[-1]))

plt.plot(result, 'hotpink')
```

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

result = []

high = []
low = []


for a in data:
    if a[-1] != "" and a[-2] != "":   # 존재하면
        if int(a[0].split('-')[0]) >= 1983: # 1983년 이후 (1983포함)
            if a[0].split('-')[1] == '02' and  a[0].split('-')[2]== '15':
                high.append(float(a[-1]))
                low.append(float(a[-2]))

plt.plot(high,'hotpink')
plt.plot(low,'skyblue')
```

## 제목 추가(한글 폰트 설정), 범레 표시

```python
import csv
import matplotlib.pyplot as plt

f = open('seoul.csv')
data = csv.reader(f)
next(data)

high = [] #최고 기온 값을 저장할 리스트 high 생성
low = [] #최저 기온 값을 저장할 리스트 low 생성

for row in data:
    if row[-1] != '' and row[-2] != "":
        date = row[0].split('-')
        if 1983 <= int(data[0]):
            if date[1] == '02' and date[2] == "15":
                high.append(float(row[-1]))
                low.append(float(row[-2]))


plt.rc('font', family='Malgun Gothic')
plt.rcParams['axes.unicode_minus'] = False
plt.title('내 생일의 기온 변화 그래프')

plt.plot(high, 'hotpink', label = '최고기온')
plt.plot(low, 'skyblue', label = '최저기온')

plt.legend()
```
