```python
import matplotlib.pyplot as plt
import csv

f=open('age.csv')
data = csv.reader(f)
result = []
name = input('인구구조가 알고싶은 지역의 이름',)
print(name)
for i in data:
    if name in i[0]:
        for a in i[3:]:
            result.append(int(a))

print(result)

plt.style.use('ggplot')
plt.rc('font', family='malgun Gothic')
plt.title(name+'지역의 인구 구조')
plt.plot(result)
```

```python
import csv
f = open('gender.csv')
data = csv.reader(f)

m = []
f = []



for i in data:
    if '신도림' in i[0]:
        for a in i[3:104]:
            m.append(-int(a))
        for a in i[106:]:
            f.append(int(a))

import matplotlib.pyplot as plt
plt.rc('font', family='malgun Gothic')
plt.title('신도림 지역의 남여 인구 분포')
plt.barh(range(101),m, label='남상') # 남성 인구수 그래프
plt.barh(range(101),f, label='여성')
plt.legend()
```

```python
import csv
import matplotlib.pyplot as plt
f = open('gender.csv')
data = csv.reader(f)
address = input('원하는 지역')
m = []
f = []
for i in data:
    if address in i[0]:
        m = i[3:104]
        f = i[106:]
plt.style.use('ggplot')
plt.figure(figsize=(10,5))
plt.rc('font', family='malgun Gothic')
plt.title(address+'지역의 남여 인구 분포')
plt.rcParams['axes.unicode_minus'] = False
plt.barh(range(101), [-int(i) for i in m] , label = '남성')
plt.barh(range(101), list(map(int,f)) , label='여성')
plt.legend()
```
