```python
import csv
f = open('gender.csv')
data = csv.reader(f)
result = []

name = input('궁금한 지역을 입력해주세요')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            result.append(int(row[i])-int(row[i+103]))
        break


import matplotlib.pyplot as plt
plt.bar(range(101),result)
plt.show()


```

```python
import csv
f = open('gender.csv')

data = csv.reader(f)

m = []
f = []

size = []

name = input('궁금한 지역을 입력해주세요')

for row in data:
    if name in row[0]:
        for i in range(3,104):
            m.append(int(row[i]))
            f.append(int(row[i+103]))

        break


import matplotlib.pyplot as plt


plt.scatter(m,f,c=range(101), alpha=0.5, cmap='jet')

plt.colorbar()

plt.plot(range(max(m)), range(max(m)),'g')
```
